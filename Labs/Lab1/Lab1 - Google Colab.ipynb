1. Введіть своє прізвище та ім'я, а також номер варіанту:

```python
student_name = "Прізвище ім'я"
variant_number = 1
```

2. Запустіть код, натисніть кнопку "Вибрати файли" та виберіть датасет, що відповідає номеру вашого варіанта

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import math

# =========================
# GOOGLE COLAB CHECK + FILE UPLOAD
# =========================

def is_colab():
    try:
        import google.colab
        return True
    except ImportError:
        return False


if is_colab():
    print("📥 Запущено в Google Colab — оберіть файл для завантаження")

    from google.colab import files
    uploaded = files.upload()

    if uploaded:
        file_name = list(uploaded.keys())[0]
        print(f"Файл завантажено: {file_name}")
    else:
        print("Файл не було обрано")

# =========================
# 1. ЗАВАНТАЖЕННЯ ДАТАСЕТУ
# =========================

if os.path.exists("/content/" + file_name):
    file_path = "/content/" + file_name
else:
    file_path = file_name

df = pd.read_csv(file_path)
```

3. Аналіз даних датасету, їхня графічна візуалізація та генерація .png-файлу

```python

# =========================
# 2. ЗАГАЛЬНА ІНФОРМАЦІЯ
# =========================
print("📊 Розмір датасету:", df.shape)
print("\n Перші 5 рядків:")
print(df.head())

print("\n Типи даних:")
print(df.dtypes)

print("\n Перевірка пропусків:")
print(df.isnull().sum())

# =========================
# 3. РОЗПОДІЛ КЛАСІВ
# =========================
class_counts = df["label"].value_counts()
print("\n Розподіл класів:")
print(class_counts)

# =========================
# 4. ВІДСОТКОВИЙ РОЗПОДІЛ КЛАСІВ
# =========================

print("\n Відсотковий розподіл класів:")
print((class_counts / len(df)) * 100)

# =========================
# 5. ДОДАТКОВИЙ АНАЛІЗ (опційно)
# =========================

print("\n Базова статистика числових ознак:")
print(df.describe())
numeric_cols = df.select_dtypes(exclude=["object"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns

# =========================
# 6. ВІЗУАЛІЗАЦІЯ ТА ЗБЕРЕЖЕННЯ
# =========================

plots = 3 + len(categorical_cols)   # 3 базові + всі categorical pie

cols = 2
rows = math.ceil(plots / cols)

fig, axs = plt.subplots(rows, cols, figsize=(16, 5 * rows))
axs = axs.flatten()
fig.suptitle(
    f"ЛР1 — Аналіз датасету\n"
    f"Датасет: {file_name}\n"
    f"Студент: {student_name} / Номер варіанту: {variant_number}",
    fontsize=16,
    y=0.98
)

idx = 0


# =========================
# 6.1. РОЗПОДІЛ КЛАСІВ
# =========================

class_counts.plot(kind='bar', ax=axs[idx])
axs[idx].set_title("Розподіл класів")
axs[idx].tick_params(axis='x', rotation=45)
axs[idx].grid(axis='y', alpha=0.3)
idx += 1

# =========================
# 6.2. ВІДСОТКИ КЛАСІВ
# =========================

percent = (class_counts / len(df)) * 100

axs[idx].pie(
    percent.values,
    labels=percent.index,
    autopct='%1.1f%%',
    startangle=90
)
axs[idx].set_title("Відсотковий розподіл класів")
idx += 1

# =========================
# 6.3. ТИПИ ОЗНАК
# =========================

sizes = [len(numeric_cols), len(categorical_cols)]
labels = ["Числові", "Текстові"]

axs[idx].pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)
axs[idx].set_title("Типи ознак")
idx += 1

# =========================
# 6.4. ДІАГРАМИ КАТЕГОРІАЛЬНИХ ОЗНАК
# =========================

for col in categorical_cols:

    value_counts = df[col].value_counts()

    axs[idx].pie(
        value_counts.values,
        labels=value_counts.index,
        autopct='%1.1f%%',
        startangle=90
    )

    axs[idx].set_title(f"{col}")
    idx += 1

# =========================
# 6.5. ВИМКНУТИ ЗАЙВІ ПОЛЯ
# =========================

for j in range(idx, len(axs)):
    axs[j].axis("off")

plt.tight_layout()
plt.subplots_adjust(top=0.93)

# =========================
# 6.6. ЗБЕРЕЖЕННЯ PNG
# =========================

script_dir = os.getcwd()
output_path = os.path.join(script_dir, f"ЛР1 - {student_name} - Варіант {variant_number}.png")
plt.savefig(output_path, dpi=300, bbox_inches="tight")

plt.show()
```
